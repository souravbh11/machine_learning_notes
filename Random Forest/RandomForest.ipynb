```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("medical_insurance.csv")
df.head()
```

```python
df["sex"].replace({"male":1,"female":0},inplace=True)
df["smoker"].replace({"yes":1,"no":0},inplace=True)
df["region"].replace({'southeast': 0, 'southwest': 1, 'northwest': 2, 'northeast': 3},inplace=True)
```

#### Train test split

```python
x = df.drop("charges",axis=1)
y = df["charges"]
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=42)
```

#### Training side

```python
rf_model = RandomForestRegressor(oob_score=True)
rf_model.fit(x_train,y_train)
```

```python
rf_model.oob_score_
```

#### Training side

```python
y_pred = rf_model.predict(x_train)
mse = mean_squared_error(y_train,y_pred)
print("MSE is",mse)
mae = mean_absolute_error(y_train,y_pred)
print("MAE is",mae)
r2 = r2_score(y_train,y_pred)
print("R2 score is",r2)
```

#### Testing side

```python
y_pred = rf_model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print("MSE is",mse)
mae = mean_absolute_error(y_test,y_pred)
print("MAE is",mae)
r2 = r2_score(y_test,y_pred)
print("R2 score is",r2)
```

#### Tuning

```python
rf_model = RandomForestRegressor(oob_score=True)

hyper_grid = {"n_estimators":np.arange(10,11),
    "criterion": ["squared_error","absolute_error"],
              "max_depth" : np.arange(3,5),
              "min_samples_split":np.arange(2,4),
              "min_samples_leaf":np.arange(2,5),
              "max_features":["sqrt","log2",1.0],
              "random_state":[45] }
gscv = GridSearchCV(rf_model,hyper_grid,cv=5)
gscv.fit(x_train,y_train)
gscv.best_estimator_

               # criterion='squared_error',
    # max_depth=None,
    # min_samples_split=2,
    # min_samples_leaf=1,
# n_estimators=100,
#     *,
#     criterion='squared_error',
#     max_depth=None,
#     min_samples_split=2,
#     min_samples_leaf=1,
# max_features =
```

#### Training side

```python
rf_model  = RandomForestRegressor(max_depth=4, min_samples_leaf=4, n_estimators=10,oob_score=True, random_state=45)
rf_model.fit(x_train,y_train)
###########################################
y_pred = rf_model.predict(x_train)
mse = mean_squared_error(y_train,y_pred)
print("MSE is",mse)
mae = mean_absolute_error(y_train,y_pred)
print("MAE is",mae)
r2 = r2_score(y_train,y_pred)
print("R2 score is",r2)
```

### Testing side

```python
y_pred = rf_model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print("MSE is",mse)
mae = mean_absolute_error(y_test,y_pred)
print("MAE is",mae)
r2 = r2_score(y_test,y_pred)
print("R2 score is",r2)
```
